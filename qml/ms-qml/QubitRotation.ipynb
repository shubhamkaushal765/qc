```python
import pennylane as qml
from pennylane import numpy as np
```

```python
dev = qml.device("lightning.qubit", wires=range(1))

@qml.qnode(dev)
def circuit(params):
    qml.RX(params[0], wires=0)
    qml.RY(params[1], wires=0)
    return qml.expval(qml.PauliZ(0))

dev_optical = qml.device("default.gaussian", wires=1)

@qml.qnode(dev_optical)
def optical_circuit(params):
    qml.Displacement(params[0], params[1], wires=0)
    qml.Kerr(params[2], wires=0)
    return qml.expval(qml.Z(0))
```

```python
params_init = 0.05 * np.random.randn(3)
params_init
```

```python
def cost(params):
    return optical_circuit(params)

optim = qml.GradientDescentOptimizer(stepsize=0.4)
params=params_init
for i in range(100):
    params, cost_ = optim.step_and_cost(cost, params)
    if i%5 == 0 or i == 99:
        print(f"Iter: {i:3d}, Cost: {cost_:.7f}")
print(params)
```
